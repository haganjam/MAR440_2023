```python
#Now trim the byfjord data
byfjord_top = byfjord_data[byfjord_data.prDM > 15]

byfjord_trim = byfjord_top[byfjord_top.prDM < 25]

#Now we trim augustas data

aug_byfjord_top = df_aug[df_aug.prDM > 15]

aug_byfjord_trim = aug_byfjord_top[aug_byfjord_top.prDM < 25]

#Now we trim the fixed data

byfjord_top_fix = byfjord_data_fix[byfjord_data_fix.prDM > 15]

byfjord_trim_fix = byfjord_top_fix[byfjord_top_fix.prDM < 25]
```

```python
#Calculate the mean of every year from the trimmed data
byfjord_mean = byfjord_trim.groupby("Year")["flECO-AFL"].mean()

byfjord_annual = pd.DataFrame(byfjord_mean).reset_index()

#Calculate the mean for augusta
aug_byfjord_mean = aug_byfjord_trim.groupby("Year")["flECO-AFL"].mean()

aug_byfjord_annual = pd.DataFrame(aug_byfjord_mean).reset_index()

#calculate the mean for fix
byfjord_mean_fix = byfjord_trim_fix.groupby("Year")["flECO-AFL"].mean()

byfjord_annual_fix = pd.DataFrame(byfjord_mean_fix).reset_index()

#Calculate trendline
z = np.polyfit(byfjord_annual["Year"], byfjord_annual["flECO-AFL"], 1)
p = np.poly1d(z)

#Augusta trendline

z1 = np.polyfit(aug_byfjord_annual["Year"], aug_byfjord_annual["flECO-AFL"], 1)
p1 = np.poly1d(z1)

#Fix trendline
z2 = np.polyfit(byfjord_annual_fix["Year"], byfjord_annual_fix["flECO-AFL"], 1)
p2 = np.poly1d(z2)
```

```python
#Now plot that
plt.rcParams["figure.figsize"]=(10,5)
plt.bar(byfjord_annual["Year"], byfjord_annual["flECO-AFL"])
plt.plot(byfjord_annual["Year"], p(byfjord_annual["Year"]) ,color="orange", linewidth="5")
plt.ylabel("Fluorescence (mg/m3)",fontweight='bold')
plt.xlabel("Year",fontweight='bold')

```
