```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import cmocean.cm as cmo
#import cartopy.crs as ccrs # these ones are not needed
#import cartopy.feature as cfeature  #these ones are not needed
import matplotlib as mpl
import gsw
import warnings
from matplotlib.lines import Line2D
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
warnings.filterwarnings("ignore")
#import OWSLib
font = {'size'   : 13}
mpl.rc('font', **font)
```

```python
station_info = pd.read_csv('byfjorden_clean_data.csv')
data_2019_ST_1 = station_info[station_info['STATION'] == "ST_1"]
data_2019_ST_2 = station_info[station_info['STATION'] == "ST_2"]
data_2019_ST_6 = station_info[station_info['STATION'] == "ST_6"]
```

```python
# calculate SA, CT and Density first (fertigstellen des Plots)
SAst9 = gsw.SA_from_SP(data_2019_ST_1["sal00"],data_2019_ST_1["prDM"],data_2019_ST_1["latitude"],data_2019_ST_1["longitude"])
CTst9 = gsw.CT_from_t(SAst9,data_2019_ST_1['t090C'],data_2019_ST_1["prDM"])
rhost9 = gsw.rho(SAst9,CTst9,data_2019_ST_1['prDM'])
data_2019_ST_1["density"]=rhost9
data_2019_ST_1
```

```python
SAst9 = gsw.SA_from_SP(data_2019_ST_2["sal00"],data_2019_ST_2["prDM"],data_2019_ST_2["latitude"],data_2019_ST_2["longitude"])
CTst9 = gsw.CT_from_t(SAst9,data_2019_ST_2['t090C'],data_2019_ST_2["prDM"])
rhost9 = gsw.rho(SAst9,CTst9,data_2019_ST_2['prDM'])
data_2019_ST_2["density"]=rhost9
data_2019_ST_2
```

```python
SAst9 = gsw.SA_from_SP(data_2019_ST_6["sal00"],data_2019_ST_6["prDM"],data_2019_ST_6["latitude"],data_2019_ST_6["longitude"])
CTst9 = gsw.CT_from_t(SAst9,data_2019_ST_6['t090C'],data_2019_ST_6["prDM"])
rhost9 = gsw.rho(SAst9,CTst9,data_2019_ST_6['prDM'])
data_2019_ST_6["density"]=rhost9
data_2019_ST_6
```

```python
#connect all the data
data_2019_ox = pd.concat([data_2019_ST_1, data_2019_ST_2, data_2019_ST_6], axis=0)
data_2019_ox
```

```python
# to convert  oxygen units from umol/kg to ml/l, umol/kg needs to be divided my the density
data_2019_ox["oxygen_ml"]=(data_2019_ox["sbox0Mm/Kg"]*data_2019_ox["density"])/44660
#data_2019_ox["oxygen_ml"].max()
data_2019_ox.to_csv('2019_fixed_oxy.csv')

```
