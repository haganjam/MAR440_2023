```python
#import everything yeah yeah yeah
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read the csvs!
df_2019 = pd.read_csv("data/2019/data_2019")
df_2020 = pd.read_csv("data/2020/data_2020")
df_2022 = pd.read_csv("data/2022/data_2022")
df_2022
```

```python
#Select the stations from 2019
st_1_19 = (df_2019.loc[df_2019['STATION']=='ST_1'])
st_2_19 = (df_2019.loc[df_2019['STATION']=='ST_2'])
st_3_19 = (df_2019.loc[df_2019['STATION']=='ST_3'])
st_4_19 = (df_2019.loc[df_2019['STATION']=='ST_4'])
st_5_19 = (df_2019.loc[df_2019['STATION']=='ST_5'])
st_6_19 = (df_2019.loc[df_2019['STATION']=='ST_6'])
```

```python
#Plot this stuff!
    #2019 plot
plt.plot(st_1_19['flECO-AFL'], st_1_19['depSM'], label="Station 1")
plt.plot(st_2_19['flECO-AFL'], st_2_19['depSM'], label="Station 2")
plt.plot(st_3_19['flECO-AFL'], st_3_19['depSM'], label="Station 3")
plt.plot(st_4_19['flECO-AFL'], st_4_19['depSM'], label="Station 4")
plt.plot(st_5_19['flECO-AFL'], st_5_19['depSM'], label="Station 5")
plt.plot(st_6_19['flECO-AFL'], st_6_19['depSM'], label="Station 6")
plt.gca().invert_yaxis()
plt.legend()
plt.ylabel('Depth (m)',fontweight="bold")
plt.xlabel('Fluorescence (mg/m3)',fontweight="bold")
plt.title("2019", fontweight="bold")
plt.show()
```
